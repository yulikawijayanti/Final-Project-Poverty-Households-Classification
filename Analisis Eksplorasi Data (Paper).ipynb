```python
# Mengimport library yang dibutuhkan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Mengatur pandas agar mengeluarkan output untuk melihat keseluruhan kolom data
pd.set_option('display.max_columns', None)
```

```python
# Mengimport data integrasi
df = pd.read_csv('data_integrasi_ta.csv', index_col='RENUM')
df.head()
```

# Jenis Lantai Rumah

```python
fig = plt.figure(figsize=(8,10))

#Bar chart persentase rumah tangga berdasarkan jenis lantai rumah
ax = fig.add_subplot(211)
ax = sns.countplot(data=df, x='Jenis lantai rumah', order=[1,2,3,4,5,6,7,8,9], ax=ax)
total = len(df['Jenis lantai rumah'])
for rec in ax.patches:
    percentage = 100 * rec.get_height()/total
    ax.text(rec.get_x() + rec.get_width() / 2, 
              rec.get_y() + rec.get_height(),
              "{:.1f}%".format(percentage),
              ha='center', 
              va='bottom')
ax.set_title('(a)')


#Bar chart persentase status kemiskinan rumah tangga berdasarkan jenis lantai rumah
ax2 = fig.add_subplot(212)
ax2 = sns.countplot(data=df, x='Jenis lantai rumah', hue='Status RT', 
                   order=[1,2,3,4,5,6,7,8,9], ax=ax2)
total2 = len(df['Jenis lantai rumah'])
for rec in ax2.patches:
    percentage = 100 * rec.get_height()/total2
    ax2.text(rec.get_x() + 0.05+ rec.get_width() / 2, 
              rec.get_y() + rec.get_height(),
              "{:.1f}%".format(percentage),
              ha='center', 
              va='bottom', size=8)
ax2.set_title('(b)')
plt.legend(('Tidak Miskin','Miskin'), loc='upper right')
plt.text(9,4000,"Keterangan:\n 1.Marmer/granit\n 2.Keramik\n 3.Parket/vinil/karpet\n 4.Ubin/tegel/teraso\n 5.Kayu/papan\n 6.Semen/bata merah\n 7.Bambu\n 8.Tanah\n 9.Lainnya")

plt.show()
```
